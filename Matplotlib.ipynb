# Lec 3 (Bar charts and Histograms)

`Bar Chart Vs Histograms`

Histograms are used to show distributions of variables while bar charts are used to compare variables. Histograms plot quantitative data with ranges of the data grouped into bins or intervals while bar charts plot categorical data (Categorical data represent characterisitcs that one can observe and sort into groups.)

## Bar Chart

```python
x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 9, 3]

x2 = [1, 3, 5, 9, 11]
y2 = [4, 3, 5, 1, 6]

plt.bar(x, y, label='Bars1', color='blue')                                  # color and label are keyword argument.
plt.bar(x2, y2, label='Bars2', color='orange')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

# Lec 5 (Stack Plots)

> Indented block



* (`composition`) A stack plot is a plot that shows the whole data set with easy visualization of how each part makes up the whole. Each constituent of the stack plot is stacked on top of each other. It shows the part makeup of the unit, as well as the whole unit.

```python
days = [1, 2, 3, 4, 5]

sleeping = [7, 8, 6, 11, 7] 
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

plt.plot([], [], color='m', label='Playing', linewidth=5)
plt.plot([], [], color='c', label='Working', linewidth=5)
plt.plot([], [], color='r', label='Eating', linewidth=5)
plt.plot([], [], color='k', label='Sleeping', linewidth=5)

plt.stackplot(days, (sleeping, eating, working, playing), colors=['m', 'c', 'r', 'k'])     # there is no label option in stackplot so we do the above trick. (creating empty lists and label them. )

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

# Lec 8 & 9 (Getting and converting data from the internet)

```python
import urllib
import matplotlib.dates as mdates
from urllib import request

def bytespdate2num(fmt, encoding='utf-8'):                              # this intire function is in doubt.
    def bytesconverter(b):
        s = b.decode(encoding)
        return (mdates.datestr2num(s))
    return bytesconverter
    

def graph_data(stock): 
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    source_code = urllib.request.urlopen(stock_price_url).read().decode() # AFTER READING THE RESPONSE ONCE IT CAN'T BE READ AGAIN (BLANK) AS PYTHON CLOSES THE CONNECTION.
    stock_data = []
    split_source = source_code.split('\n')
    for line in split_source[1:]:
        split_line = line.split(',')
        if len(split_line) == 7:
            if 'values' not in line and 'labels' not in line:
                stock_data.append(line)

    date, closep, highp, lowp, openp, adj_closep, volume = np.loadtxt(stock_data,
                                                          delimiter=',',
                                                          unpack=True,   # we specify that we indeed want to unpack the variables here not just to one variable, but to this group of variables we've defined
                                                          # %Y = full year. 2015
                                                          # %y = partial year 15
                                                          # %m = number month
                                                          # %d = number day
                                                          # %H = hours
                                                          # %M = minutes
                                                          # %S = seconds
                                                          # 12-06-2014
                                                          # %m-%d-%Y
                                                          converters={0: bytespdate2num('%Y-%m-%d')}) # converters for converting data to make it recognisable by matplotlib.
                                                                                                      # we use the optional "converters" parameter to specify what element we want to convert (0), and then how we want to do that.

    plt.plot_date(date, closep,'-', label='Price')
 
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Interesting Graph\nCheck it out')
    plt.legend()
    plt.show()


graph_data('TSLA')
```

```python
import matplotlib.ticker as mticker
from mplfinance.original_flavor import candlestick_ohlc     # To access the old API with the new mplfinance package installed
```

```python
def graph_data(stock): 

    fig = plt.figure()                                 
    ax1 = plt.subplot2grid((1, 1), (0, 0))              
                                                        
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    source_code = urllib.request.urlopen(stock_price_url).read().decode()           
    stock_data = []
    split_source = source_code.split('\n')
    for line in split_source[1:]:
        split_line = line.split(',')
        if len(split_line) == 7:
            if 'values' not in line and 'labels' not in line:
                stock_data.append(line)

    date, closep, highp, lowp, openp, adj_closep, volume = np.loadtxt(stock_data,
                                                          delimiter=',',
                                                          unpack=True,                                                        
                                                          converters={0: bytespdate2num('%Y-%m-%d')})
    x = 0
    y = len(date)
    ohlc = []
    
    while (x < 30):                                                                 # x < y is using data of 10 years but it made the graph clumsy so we consider only for a month (30 days).
        append_me = date[x], openp[x], highp[x], lowp[x], closep[x], volume[x]
        ohlc.append(append_me)
        x += 1

    candlestick_ohlc(ax1, ohlc, width=0.4, colorup='g', colordown='r')
    
    for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(45)
    
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
    ax1.xaxis.set_major_locator(mticker.MaxNLocator(10))                                    # it's creating 9 more date values in x-axis.
    ax1.grid(True)

    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(stock)
    # plt.legend()
    plt.subplots_adjust(left=0.09, bottom=0.20, right=0.94, top=0.90, wspace=0.2, hspace=0)       
    plt.show()

graph_data('TSLA')
```
