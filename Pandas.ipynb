`QUARTILE`: In statistics, a quartile is a type of [quantile](https://en.wikipedia.org/wiki/Quantile) which divides the number of data points into four parts, or quarters, of more-or-less equal size. The data must be ordered from smallest to largest to compute quartiles; as such, quartiles are a form of order statistic. The three main quartiles are as follows:

- The first quartile (Q1) is defined as the middle number between the smallest number (minimum) and the median of the data set. It is also known as the lower or 25th empirical quartile, as 25% of the data is below this point.
- The second quartile (Q2) is the median of a data set; thus 50% of the data lies below this point.
- The third quartile (Q3) is the middle value between the median and the highest value (maximum) of the data set. It is known as the upper or 75th empirical quartile, as 75% of the data lies below this point.

`What is the 25th 50th and 75th percentile?`
- Every score is in the 100th percentile. The 25th percentile is also known as the first quartile (Q1), the 50th percentile as the median or second quartile (Q2), and the 75th percentile as the third quartile (Q3). In general, percentiles and quartiles are specific types of quantiles.

```python
# d_parser = lambda x: datetime.strptime(x, '%Y-%m-%d %I-%p')
td = pd.read_csv('/content/drive/MyDrive/ETH_1h.csv') #, parse_dates = ['Date'], date_parser = d_parser )      
                                                              # we need to create a function that converts each string to datetime.
td.head()
```

- Datetime Formatting Codes - http://bit.ly/python-dt-fmt

```python
td['Date'] = pd.to_datetime(td['Date'], format = '%Y-%m-%d %I-%p')  # go to the read_csv line for loading it at the file reading time.
td.dtypes
```

```python
td['Date'].head()
```

```python
td.loc[0, 'Date'].day_name()
```

```python
# accessing the 'dt class' from the 'series object' to apply the above function for the entire 'Date' column
td['Date'].dt.day_name()
```

```python
td['DayOfWeek'] = td['Date'].dt.day_name()
td
```

```python
# earliest datetime
td['Date'].min()
```

```python
# latest datetime
td['Date'].max()
```

```python
# time delta - the amount of time that spans between the above dates or (the diff. of two dates.)
td['Date'].max() - td['Date'].min()
```

```python
filt = (td['Date'] >= '2019') & (td['Date'] < '2020')                      # if pandas is able to identify the format then no need to convert it using to_datetime.
td.loc[filt]
```

```python
filt = (td['Date'] >= pd.to_datetime('2019-01-01')) & (td['Date'] < pd.to_datetime('2020-01-01'))
td.loc[filt]
```

```python
td.set_index('Date', inplace=True)
```

```python
td['2020']                      # as date is set to index we can directly access the dates.
```

```python
# using slice for a range of data based on dates.
td['2020-01':'2020-02']                                                     # the 2nd date is inclusive value.
```

```python
# average closing value for a particular time frame. (here 2 months)
td['2020-01':'2020-02']['Close'].mean()
```

```python
# highest value for a day
td['2020-01-01']['High'].max()
```

`Resampling`
- Sampling is an active process of gathering observations with the intent of estimating a population variable. Resampling is a methodology of economically using a data sample to improve the accuracy and quantify the uncertainty of a population parameter

```python
highs = td['High'].resample('D').max()                # first is was hourly, D will make it on a daily basis. (click on the below click for more)
highs['2020-01-01']
```

- Pandas Date Offset Codes - http://bit.ly/pandas-dt-fmt

```python
import matplotlib.pyplot as plt
highs.plot()
```

```python
td.resample('W').mean()                         # if we need the same aggregation method on every column.
```

```python
# using agg() for applying diff. aggretation methods on diff. columns
dict = {
    'Close' : 'mean',
    'High' : 'max',
    'Low' : 'min', 
    'Volume' : 'sum',
}
td.resample('W').agg(dict)
```
